```python
#展示y=x^(4/3)在[-3,3]的值
import numpy as np
import matplotlib.pyplot as plt

f = lambda x:np.cbrt(x ** 4)
x = np.linspace(-3.0,3.0,100)
y = np.array([f(x0) for x0 in x])
plt.plot(x,y)
plt.show()
```

```python
#用二阶牛顿法求解（可以发现无法收敛）
import random

x0 = random.uniform(-3,3)
print("起始点x={}".format(x0))

max_step = 100
eps = 1e-7
g = lambda x1: 4/3 * np.cbrt(x1)
g2  = lambda x1:4/9 * (1 / np.cbrt(x1 * x1))

for i in range(max_step):
    x1 = x0 - g2(x0) / g(x0) 
    if abs(x0 - x1) < eps:
        x0 = x1
        break
    x0 = x1

print("使得y取得最小值的x={}".format(x0))
```

```python
#用一阶牛顿法求解
x0 = random.uniform(-3,3)
print("起始点x={}".format(x0))

#记录求解过程
xx,yy,dx,dy = [],[],[],[]
for i in range(max_step):
    x1 = x0 - f(x0) / g(x0) 
    if abs(x0 - x1) < eps:
        x0 = x1
        break
    dx.append(- f(x0) / g(x0))
    xx.append(x0)
    yy.append(f(x0))
    dy.append(f(x1) - f(x0))
    x0 = x1

print("使得y取得最小值的x={}".format(x0))
```

```python
#结果可视化
plt.plot(x,y)
plt.plot(x0,f(x0),marker = 'o',color = 'orange')
#绘制求解过程图
plt.quiver(xx,yy,dx,dy,angles = 'xy', scale_units = 'xy',scale = 1,width = 0.005,color = 'red')
plt.show()
```
